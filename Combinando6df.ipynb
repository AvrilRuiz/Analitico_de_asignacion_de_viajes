```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_pie_with_autofontsize(sizes, ax=None, autopct='%1.1f%%', startangle=90, legend=True, labels=None, fontsize=8):
    if ax is None:
        fig, ax = plt.subplots()
    wedges, texts, autotexts = ax.pie(sizes, autopct=autopct, startangle=startangle, labels=labels, textprops={'fontsize': fontsize})
    for i, autotext in enumerate(autotexts):
        size = sizes[i]
        percentage = size / sum(sizes) * 100
        if percentage >= 10:
            fontsize = 12
        elif percentage >= 5:
            fontsize = 10
        elif percentage >= 3:
            fontsize = 8    
        else:
            fontsize = 5
        autotext.set_fontsize(fontsize)
    ax.axis('equal')
    return ax
```

## Exploración cuantitativa  

```python
cuantitativas = ['CANT_PROGRAMADA', 'PESO_NETO', 'CAP_MAXIMA']
```

```python
 df[cuantitativas].describe()
```

```python
estadisticas_cuantitativas = pd.DataFrame(columns=['variable', 'promedio', 'mediana', 'moda', 'maximo', 'minimo', 'rango', 'varianza', 'desviacion_estandar'])

for column in cuantitativas: 
  df_variable = df[column]

  # Calcular medidas de tendencia central
  promedio = df_variable.mean()
  mediana = df_variable.median()
  moda_ = df_variable.mode()

  if len(moda_) == 1:
    moda = moda_[0]
  else:
    moda = [moda_[i] for i in range(len(moda_))]

  # Calcular medidas de dispersión
  maximo = df_variable.max()
  minimo = df_variable.min()
  rango = df_variable.max() - df_variable.min()
  varianza = df_variable.var()
  desviacion_estandar = df_variable.std()

  # Agregar a un nuevo data frame
  nuevo_registro = {'variable': column, 'promedio': promedio, 'mediana': mediana, 'moda': moda, 'maximo': maximo,
                    'minimo': minimo, 'rango': rango, 'varianza': varianza, 'desviacion_estandar': desviacion_estandar}
  estadisticas_cuantitativas = pd.concat([estadisticas_cuantitativas, pd.DataFrame([nuevo_registro])], ignore_index=True)
```
